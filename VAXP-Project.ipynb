```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("llmdata.csv")

df.head()
```

```python
df = df.dropna(subset=["provider"])
state_counts = df["provider"].value_counts().head(10)

plt.figure()
state_counts.sort_values().plot(kind='barh')
plt.xlabel("# of models")
plt.ylabel("Providers")
plt.title("Top 10 Providers by # of Models")
plt.show()
```

```python
box_df = df.dropna(subset=["pricing_tier", "aa_intelligence_index"])

tier_order = ["Free", "Budget", "Mid", "Premium", "Ultra", "Unknown"]

# Create box plot
plt.figure(figsize=(10,6))
box_df.boxplot(column="aa_intelligence_index", by="pricing_tier", grid=False)

plt.title("Intelligence Distribution by Pricing Tier")
plt.suptitle("")
plt.xlabel("Pricing Tier")
plt.ylabel("AA Intelligence Index")
plt.show()
```

```python
#3 Find top-5 models on 5 benchmarks  (GPQA Diamond','Humanity\'s\nLast Exam','LiveCode\nBench','SciCode','MATH-500')

#GPQA Diamond
print(f"\nTop 5 models for gpqa_diamond:")
top5_gpqa = df[["model_name", "provider", "gpqa_diamond"]]
top5_gpqa = top5_gpqa.dropna(subset=["gpqa_diamond"])
top5_gpqa = top5_gpqa.sort_values(by="gpqa_diamond", ascending=False)
top5_gpqa = top5_gpqa.head(5)
print(top5_gpqa)

#Humanitys last exam
print(f"\nTop 5 models for humanitys_last_exam:")
top5_hle = df[["model_name", "provider", "humanitys_last_exam"]]
top5_hle = top5_hle.dropna(subset=["humanitys_last_exam"])
top5_hle = top5_hle.sort_values(by="humanitys_last_exam", ascending=False)
top5_hle = top5_hle.head(5)
print(top5_hle)
#livecodebench
print(f"\nTop 5 models for livecodebench:")
top5_livecode = df[["model_name", "provider", "livecodebench"]]
top5_livecode = top5_livecode.dropna(subset=["livecodebench"])
top5_livecode = top5_livecode.sort_values(by="livecodebench", ascending=False)
top5_livecode = top5_livecode.head(5)
print(top5_livecode)

#scicode
print(f"\nTop 5 models for scicode:")
top5_scicode = df[["model_name", "provider", "scicode"]]
top5_scicode = top5_scicode.dropna(subset=["scicode"])
top5_scicode = top5_scicode.sort_values(by="scicode", ascending=False)
top5_scicode = top5_scicode.head(5)
print(top5_scicode)

#math_500
print(f"\nTop 5 models for math_500:")
top5_math = df[["model_name", "provider", "math_500"]]
top5_math = top5_math.dropna(subset=["math_500"])
top5_math = top5_math.sort_values(by="math_500", ascending=False)
top5_math = top5_math.head(5)
print(top5_math)
```
